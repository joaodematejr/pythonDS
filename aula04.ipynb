```python
#Crescimento da população brasileira 1980 - 2016
#DataSus
import matplotlib.pyplot as plt
dados = open("original.csv").readlines()

#Anos
x = []
#População
y = []

for i in range(len(dados)):
    if i != 0:
        linha = dados[i].split(";")
        x.append(int(linha[0]))
        y.append(int(linha[1]))

plt.bar(x, y, color="#E4E4E4")
plt.plot(x, y, color="k", linestyle="--")
#plt.scatter(x, y, color="k")
plt.title("Crescimento da população brasileira 1980 - 2016", color='w')
plt.xlabel("Ano", color='w')
plt.ylabel("População x100.000.000", color='w')
#plt.show()
plt.savefig("populacao_brasileiro.png", dpi=300)
```
